```python
def get_seqs(path, fns):
    seqs = []
    pattern = re.compile('[a-zA-Z]')
    for fn in fns:
        if not bool(pattern.search(fn.split('.')[0].split('/')[-1])) and fn.split('.')[-1] == 'txt':
            seq = []
            with open(path+fn) as f:
                for l in f:
                    l = l.strip()
                    l = re.sub(r'\s+', ' ', l)
                    ls = l.split(' ')
                    if len(ls) >= 5:
                        seq.append(ls[5].strip())
            seqs.append(' '.join(seq))
    return seqs
```
